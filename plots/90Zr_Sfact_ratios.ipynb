```python
import matplotlib.pyplot as plt

from matplotlib import rc
rc('text',usetex=True)
rc('font',**{'family':'Times New Roman'})

# plt.rc('axes', labelsize=35)
# plt.rcParams["font.family"] = "Times New Roman"
from matplotlib.ticker import (MultipleLocator, FormatStrFormatter, AutoMinorLocator)
```

```python
x = ["1/2-", "9/2+", "3/2-", "5/2-"]
emd28 = [0.56, 0.51, 0.56, 0.55]
em = [0.67, 0.19, 0.64, 0.61]
emd28err = [0.07/1.2756, 0.05/1.0629, 0.14/3.3022, 0.19/5.0002]
emerr = [0.07/1.0671, 0.05/2.8405 , 0.14/2.9290, 0.19/4.5142 ]
f = plt.figure()
f.set_figwidth(10)
f.set_figheight(10)

plt.plot(x, emd28, color='blue', linestyle='solid', marker='o', markerfacecolor='blue', markersize=12, label='n50emd28')
plt.plot(x, em, color='orange', linestyle='solid', marker='o', markerfacecolor='orange', markersize=12, label='n50em')
plt.errorbar(x, emd28, yerr=emd28err, capsize=8)
plt.errorbar(x, em, yerr=emerr, capsize=8)

plt.xlabel('J$\pi$ State of $^{89}$Y', fontsize=35)
plt.xticks(fontsize=25)
plt.yticks(fontsize=25)
plt.ylabel('R = S$_{\\rm exp}$/C$^2$S', fontsize=35)
plt.title('$^{90}$Zr(e,e\'p)',fontsize=35)

plt.minorticks_on()
plt.tick_params(which="both", direction="in")
plt.ylim(0.0, 0.85)
plt.legend(loc=4,fontsize=25)
plt.show()
```
