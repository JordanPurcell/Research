```python
import matplotlib.pyplot as plt
import numpy as np
import math

plt.rc('axes', labelsize=25)
plt.rc('xtick', labelsize=20)
plt.rc('ytick', labelsize=20)
#plt.rcParams['text.usetex'] = True
plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams['axes.linewidth'] = 3.0

A = [80, 82, 84, 86, 88, 90, 92, 94, 96, 98]                           # Nucleon numbers

mu_exp = [2.20, 4.12, 2.75131, 2.44, -0.1374154, 6.37, 2.5, 6.259, 3.0, 10.84, 6.521, -0.101, 8.14, 10.82, 2.3, 11.30, 6.32, 10.46, 10.97]
mu_err = [0.1, 0.56, 0.12, 0.22, 0.0000003, 0.04, 0.4, 0.013, 0.2, 0.06, 0.002, 0.002, 0.13, 0.14, 0.3, 0.05, 0.06, 0.05, 0.06]
mu_th = [2.65676, 4.25305, 2.76323, 2.27189, -0.02839, 6.63283, 2.79427, 6.58524, 3.87149, 11.7607, 6.7089, -0.03349 , 8.8197, 11.7316, 2.93701, 11.9339, 6.72065, 11.5445, 11.98071]

# Q1 calculated with e_p = 1.8e, Q2 calculated with e_p = 2e
# consider Q3 with e_p = 1.72e
Qexp = [13.2, -43, -51, -25, -35]
Q_err = [0.9, 6, 3, 3, 2]
Q2 = [20.45, -45.13, -61.37, -30.28, -39.79]
Q1 = [0.9*Q2[i] for i in range(len(Q2))]
Qfree = [0.5*Q2[i] for i in range(len(Q2))]

BE2_exp = [730, 1210, 1050, 1060, 903, 627, 975]                           # Experimental BE2 for first 0+ -> 2+ transitions for 80Zn - 92Mo
BE2_err_top = [90, 150, 150, 100, 32, 34, 43]
BE2_err_bot = [90, 150, 150, 100, 23, 34, 43]
BE2_th1 = [608, 928, 952, 985, 941, 612, 996, 1265, 1174, 751]             # Calculations for same transitions using e_p = 1.8e, values after 996 correspond to 94Ru, 96Pd, and 98Cd 
BE2_th2 = [751, 1146, 1176, 1217, 1162, 755, 1230]                         # Calculations for same transitions using e_p = 2e, values after 1230 correspond to 94Ru, 96Pd, and 98Cd 
BE2_JW = [604, 918, 1185, 833, 691, 561, 1097, 1334, 1235, 809]            # Ji and Wildenthal calculations for first 0+ -> 2+ transitions (e_p = 2.0e)

# 94Rb only has J values of 6 and 8+
# M = sqrt[B(E2)]
M_exp = [math.sqrt(BE2_exp[i]) for i in range(len(BE2_exp))]
M_err_top = [0.5*BE2_err_top[i]*math.pow(BE2_exp[i], -0.5) for i in range(len(BE2_exp))]
M_err_bot = [0.5*BE2_err_bot[i]*math.pow(BE2_exp[i], -0.5) for i in range(len(BE2_exp))]
M_th1 = [math.sqrt(BE2_th1[i]) for i in range(len(BE2_exp))]
M_th2 = [math.sqrt(BE2_th2[i]) for i in range(len(BE2_exp))]
M_free = [0.5*M_th2[i] for i in range(len(M_th2))]
x = np.linspace(-65, 38)

plt.figure(figsize=(10,7))
plt.plot(x, x, 'k--')
plt.errorbar(Q1, Qexp, yerr=Q_err, fmt='o', mfc='orange', mec='orange', ecolor='orange', label='Q')
plt.errorbar(M_th1, M_exp, yerr=(M_err_bot, M_err_top), fmt='*', mfc='purple', mec='purple', ecolor='purple', ms=10, label=r'M (0$^+_1 \rightarrow$ 2$^+_1$)')
plt.vlines(0, -65, 38, colors='k')
plt.hlines(0, -65, 38, colors='k')
plt.xlim(-65, 38)
plt.ylim(-65, 38)
plt.xlabel(r'Theory (e$ \dot $fm$^2$)', fontsize=20)
plt.ylabel(r'Exp (e$ \dot $fm$^2$)', fontsize=20)
plt.tick_params(which="major", direction="in", right=True, top=True, length=10, width=2, labelsize=20)
plt.tick_params(which="minor", direction="in", right=True, top=True, length=5, width=1, labelsize=20)
plt.title(r'Q and M (e$_\mathrm{p}$ = 1.8e)', fontsize=25)
plt.legend(fontsize=20)
plt.show()

plt.figure(figsize=(10,7))
plt.plot(x, x, 'k--')
plt.errorbar(Qfree, Qexp, yerr=Q_err, fmt='o', mfc='orange', mec='orange', ecolor='orange', label='Q')
plt.errorbar(M_free, M_exp, yerr=(M_err_bot, M_err_top), fmt='*', mfc='purple', mec='purple', ecolor='purple', ms=10, label=r'M (0$^+_1 \rightarrow$ 2$^+_1$)')
plt.vlines(0, -65, 38, colors='k')
plt.hlines(0, -65, 38, colors='k')
plt.xlim(-65, 38)
plt.ylim(-65, 38)
plt.xlabel(r'Theory (e$ \dot $fm$^2$)', fontsize=20)
plt.ylabel(r'Exp (e$ \dot $fm$^2$)', fontsize=20)
plt.tick_params(which="major", direction="in", right=True, top=True, length=10, width=2, labelsize=20)
plt.tick_params(which="minor", direction="in", right=True, top=True, length=5, width=1, labelsize=20)
plt.title(r'Q and M (e$_\mathrm{p}$ = 1.0e)', fontsize=25)
plt.legend(fontsize=20)
plt.show()

plt.figure(figsize=(10,7))
plt.errorbar(M_th1, M_exp, yerr=(M_err_bot, M_err_top), fmt='*', mfc='purple', mec='purple', ecolor='purple', ms=10, label='M (e$_\mathrm{p}$ = 1.8e)')
plt.errorbar(M_th2, M_exp, yerr=(M_err_bot, M_err_top), fmt='*', mfc='red', mec='red', ecolor='red', ms=10, label='M (e$_\mathrm{p}$ = 2.0e')
plt.plot(x, x, 'k--')
plt.xlim(20, 38)
plt.ylim(20, 38)
plt.xlabel(r'Theory (e$ \dot $fm$^2$)', fontsize=20)
plt.ylabel(r'Exp (e$ \dot $fm$^2$)', fontsize=20)
plt.tick_params(which="major", direction="in", right=True, top=True, length=10, width=2, labelsize=20)
plt.tick_params(which="minor", direction="in", right=True, top=True, length=5, width=1, labelsize=20)
plt.title(r'M', fontsize=25)
plt.legend(fontsize=20)
plt.show()

plt.figure(figsize=(10,7))
plt.errorbar(mu_th, mu_exp, yerr=mu_err, fmt='o')
plt.plot(x, x, 'k--')
plt.vlines(0, -65, 38, colors='k')
plt.hlines(0, -65, 38, colors='k')
plt.xlim(-1, 13)
plt.ylim(-1, 13)
plt.xlabel(r'Theory ($\mathrm{\mu}_\mathrm{N}$)', fontsize=20)
plt.ylabel(r'Exp ($\mathrm{\mu}_\mathrm{N}$)', fontsize=20)
plt.tick_params(which="major", direction="in", right=True, top=True, length=10, width=2, labelsize=20)
plt.tick_params(which="minor", direction="in", right=True, top=True, length=5, width=1, labelsize=20)
plt.title('Dipole Moments', fontsize=25)
plt.show()

fig4 = plt.figure(figsize=(10,7))
ax4 = fig4.add_subplot()
plt.errorbar(A[:7], BE2_exp, yerr=(BE2_err_bot, BE2_err_top), fmt='x', mfc='purple', mec='purple', ecolor='purple', ms=10, label='exp')
plt.plot(A, BE2_th1, 'ko-', label='r35c')
plt.plot(A, [0.81*BE2_JW[i] for i in range(len(BE2_JW))], '-o', label='JW')
plt.xlim(78, 100)
plt.ylim(0, 1500)
plt.xticks(A)
plt.xlabel('A', fontsize=20)
plt.ylabel('B(E2) e$^2 \dot $fm$^4$', fontsize=20)
plt.tick_params(which="major", direction="in", right=True, top=True, length=10, width=2, labelsize=20)
plt.tick_params(which="minor", direction="in", right=True, top=True, length=5, width=1, labelsize=20)
plt.legend(fontsize=20)
ax4.text(0.95, 0.05, r'0$^+_1$ $\rightarrow$ 2$^+_1$ for N=50' + '\n' + r'e$_\mathrm{p}$ = 1.8e and skx radial wavefunctions',
        verticalalignment='bottom', horizontalalignment='right',
        transform=ax4.transAxes,
        color='black', fontsize=20)
plt.show()
```
